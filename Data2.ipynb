```python
#Create an app that stores the data of customers inside a python class
from unicodedata import name
import matplotlib as plt
import seaborn as sns 
import pandas as pd
import numpy as np

class Customer:

    def __init__(self, name, ad, con_num, email):
        self.name = name
        self.ad = ad
        self.con_num = con_num
        self.email = email
        self.unpack()

    def unpack(self):
        self.df = pd.read_csv(r"CustomerInfo.csv")
        self.df = self.df.drop(columns=self.df.columns[0])
        print(self.df)
    
    def store(self):
        data = [self.name,self.ad,self.con_num,self.email]
        self.df.loc[self.df.shape[0]] = data
        self.df.to_csv(r"CustomerInfo.csv")

name = input("What is your name?")
ad = input("What is your address?")
con_num = (input("What is your mobile number or landline?"))
email = input("What is you email address?")

customer1 = Customer(name,ad,con_num,email)
customer1.store()
```
