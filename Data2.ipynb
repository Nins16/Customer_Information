```python
#Create an app that stores the data of customers inside a python class
from unicodedata import name
import matplotlib as plt
import seaborn as sns 
import pandas as pd
import numpy as np
import csv 

class customer:

    def __init__(self, name, ad, con_num, email):
        dict = {}
        self.name = name
        self.ad = ad
        self.con_num = con_num
        self.email = email
    
        def unpack_csv():
            csv_filename = 'CustomerInfo.csv'
            with open(csv_filename) as f:
                reader = csv.DictReader(f)

    
    def store():
        dict = [{'Name':name,
               'Address':ad,
               'Contact':con_num,
                "Email Address":email
                }]
        headers = ['Name','Address','Contact','Email Address']
        with open("CustomerInfo.csv","w",newline = "") as csv_file:
            writer = csv.DictWriter(csv_file,fieldnames=headers)
            writer.writeheader()
            for data in dict:
                writer.writerow(data)              

        csv_file.close()      

name = input("What is your name?")
ad = input("What is your address?")
con_num = input("What is your mobile number or landline?")
email = input("What is you email address?")

customer.store()
```
